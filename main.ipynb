```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# plt.style.available

# plt.style.use('')
```

```python
# ##### the relation between height and weight.
df = pd.read_csv('lifestyle.csv')
df = df.head(50)
df
```

```python
###### the relation between height and weight.
fig,axe = plt.subplots()

plt.figure(figsize=(8,4),dpi=300)
axe.scatter(df['Height (m)'],df['Weight (kg)'])


axe.set_title('relation between height and weight.')
axe.set_xlabel('Height')
axe.set_ylabel('Weight')
axe.grid(linestyle='dashed')
axe.spines['top'].set_visible(False)
axe.spines['left'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
###### which workout type is burn more calorie.
sns.barplot(x ='Workout_Type',y  = 'Burns Calories (per 30 min)' ,data=df)

plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')

plt.show()
```

```python
###### which workout takes less BPM

sns.barplot(x='Workout_Type',y ='Avg_BPM', data = df)

plt.grid(linestyle='--')
```

```python
###### which gender is more rating
sns.barplot(x ='Gender',y = 'rating',data = df)
```

```python
###### which gender have more lean mass
sns.swarmplot(x='lean_mass_kg',y='Gender',data=df)
```

```python
###### corelation between rating and rep,set,calories burned

core =df[['Reps','rating','Sets','Burns Calories (per 30 min)']].corr()

sns.heatmap(core,annot=True)
```

```python
###### cooking method and prep_time

sns.lineplot(x='cooking_method',y='prep_time_min',data =df)
plt.grid(linestyle='--')
```

```python
###### which excercise have less set
sns.barplot(x ='Workout_Type',y='Sets',data = df,palette='plasma') 
plt.grid()
plt.show()
```

```python
###### relation bw bpm and workout type

sns.barplot(x = 'Avg_BPM', y = 'Workout_Type', data= df)
```

```python
######  relation bw workout and water intake

sns.barplot(x = 'Workout_Type',y='Water_Intake (liters)',data = df)
```

```python
###### relation bw water intake and ggender

sns.boxplot(x = 'Gender',y= 'Water_Intake (liters)',data=df,palette='inferno')
```

```python
###### relation bw fat% and diet type

sns.barplot(x = 'diet_type',y= 'Fats',data = df)
```

```python
###### relation bw sugar% and diet type
sns.barplot(x = 'diet_type',y= 'sugar_g',data = df)
```

```python
###### relation bw fat% and sugar
core =df[['sugar_g','Fats']].corr()
sns.heatmap(core,annot=True)
```

```python
###### age bpm
sns.kdeplot(y = 'Avg_BPM',x = 'Age', data = df,color = 'red')
plt.grid()
plt.tight_layout()
```

```python
###### age fat%


sns.kdeplot(y = 'Fats',x = 'Age',data = df)
plt.grid()
plt.show()
```

```python
###### workout type and protien.

sns.lineplot(x='Workout_Type',y = 'Proteins',data = df)
plt.grid(linestyle='--')
plt.show()
```
